# VALOR PRESENTE NETO

# El _Valor Presente Neto_ **(VPN)**, también llamado _Valor Actual Neto_ **(VAN)**, es una métrica financiera que calcula la rentabilidad de una inversión trayendo flujos de efectivo futuros al presente y restando la inversión inicial. Determina si un proyecto generará valor ***(VPN > 0)*** o lo destruirá ***(VPN < 0)***, considerando el costo del dinero en el tiempo y la tasa de descuento.

$$ VPN = \frac{VF}{(1 + i)^n} $$
